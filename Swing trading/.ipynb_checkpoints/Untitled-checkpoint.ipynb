```python
import requests
import pandas as pd
```

```python
headers={ 'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'}
session=requests.session()
session.get("http://www1.nseindia.com", headers = headers)
```

```python
def Volume(symbol,Time,dataType):
    url="https://www1.nseindia.com/products/dynaContent/common/productsSymbolMapping.jsp?symbol="+symbol+"&segmentLink=3&symbolCount=2&series=ALL&dateRange="+Time+"&fromDate=&toDate=&dataType="+dataType
    data = session.get(url, headers =headers)
    df=pd.read_html(data.content)[0]
    return df
```
