```python
df_data_n=df_data[df_data["PE.Expiry"]=="2020-11-05"]
df_data_n
```

```python
df_data_n["Difference"]=df_data_n["CE.underlyingValue"]-df_data_n["CE.strikePrice"]
df_data_n["Ratio"]=df_data_n["CE.openInterest"]/df_data_n["PE.openInterest"]
```

```python
df_call=df_data_n[df_data_n["Difference"]<0].iloc[3:]
df_put=df_data_n[df_data_n["Difference"]>0].iloc[:-3]
```

```python
df_call_1=df_call.sort_values(by=["CE.openInterest"],ascending=False).iloc[:7].sort_values(by=['Ratio'],ascending=False)
df_put_1=df_put.sort_values(by=["PE.openInterest"],ascending=False).iloc[:7].sort_values(by=['Ratio'])
```

```python
def OverView(CE_Sell,PE_Sell,CE_Buy,PE_Buy):
    Max_Profit=round((CE_Sell+PE_Sell-CE_Buy-PE_Buy)*50,2)
    Max_Loss=round((CE_Sell+PE_Sell-CE_Buy-PE_Buy-200)*50,2)
    return [Max_Profit,Max_Loss,round(-Max_Loss/Max_Profit,2)]
    
```

```python
def CallPrice(StrikePrice,Data):
    return Data[Data["CE.strikePrice"]==StrikePrice]["CE.lastPrice"].iloc[0]

def PutPrice(StrikePrice,Data):
    return Data[Data["PE.strikePrice"]==StrikePrice]["PE.lastPrice"].iloc[0]
```

```python
my_columns = ["CE StrikePrice", 'PE StrikePrice',"CE StrikePrice 1", 'PE StrikePrice 1','Max Profit', 'Max Loss', "Risk/Reward"]
final_dataframe = pd.DataFrame(columns = my_columns)


for i in df_call_1["CE.strikePrice"]:
    l=[]
    l.append(i)
    for j in df_put_1["PE.strikePrice"]:
        l.append(j)
        l.append(l[0]+200)
        l.append(l[1]-200)
        CE_Sell=CallPrice(l[0],df_data_n)
        PE_Sell=PutPrice(l[1],df_data_n)
        CE_Buy=CallPrice(l[2],df_data_n)
        PE_Buy=PutPrice(l[3],df_data_n)
        l=l+OverView(CE_Sell,PE_Sell,CE_Buy,PE_Buy)
        final_dataframe = final_dataframe.append(
                                        pd.Series(l,index = my_columns), 
                                        ignore_index = True)
        l=[i]
        
```

```python
final_dataframe=final_dataframe[final_dataframe["Risk/Reward"]<=3]
final_dataframe
```

```python
final_dataframe["Diff"]=final_dataframe["CE StrikePrice"]-final_dataframe['PE StrikePrice']
```
